💙 **np.random.rand**

Random numbers between 0 and 1 (uniform distribution)

````python
np.random.rand(3,4)
np.random.rand(3)
````

💙 **np.random.randn**

Random Numbers with mean 0 and variance 1 (standard normal distribution)

````python
np.random.randn(7,4)
np.random.randn(2)
````

💙 **np.random.randint**

Random integer between [low, high) (meaning high is not included)

````python
np.random.randint(10)
np.random.randint(1,30)
np.random.randint(1,30,10)
np.random.randint(1,30,(3,5))
````

💙 **np.random.normal(loc=0.0, scale=1.0, size=None)**

loc = mean <br>
scale = standard deviation

````python
np.random.normal(10, 2 , 5)
````

💙 **np.random.random(size=None)**

Returns random floats in the half-open interval [0.0, 1.0]

````python
print(np.random.random() ,
np.random.random(5),
np.random.random((3, 3)),
np.random.random(5) * 10)
````

💙 **np.random.uniform(low=0.0, high=1.0, size=None)**

This function is used to generate random numbers from a continuous uniform distribution.
That is, all numbers between low and high are selected with equal probability
"rand is only for (0,1).<br>
uniform is more flexible and allows specifying any desired interval

````python
np.random.uniform()
np.random.uniform(2,7)
np.random.uniform(2,7,10)
np.random.uniform(2,7,(3,3))
````

💙 **np.random.choice(a, size=None, replace=True, p=None)**

Random selection from the elements of an array 

Parameters:

a → array or number of elements (0 to n-1) <br>
size → number of samples <br>
replace → whether to sample with replacement or not <br>
p → array of probabilities for each element (sum must = 1)

````python
np.random.choice([1,2,3,4,5])

arr = np.array([1,2,3,4,5])
np.random.choice(arr)
np.random.choice(arr, 3) # With repetition
np.random.choice(arr, 3, replace=False) # The argument p is a list of probabilities that specifies the probability of selecting each element from the array arr.
````

💙 **np.random.shuffle**

Randomly shuffles the order of the elements

````python
arr = np.arange(10)
np.random.shuffle(arr)
arr 
````

💙 **np.random.permutation**

Unlike np.random.shuffle(), which modifies the array in-place, np.random.permutation() returns a new array

````python
arr = np.arange(10)
p = np.random.permutation(arr)
print("Original:", arr)
print("Permutation:", p)
````

This cheat sheet includes commonly used NumPy functions for array manipulation.

```python
import numpy as np

a = np.array([[1, 2, 3],
              [4, 5, 6]])
 

# 1️⃣ reshape
# فارسی: تغییر شکل آرایه بدون تغییر داده‌ها
# English: Changes the shape of an array without modifying the data
b = a.reshape(3, 2)
 

# 2️⃣ flatten
# فارسی: تبدیل آرایه به 1بعدی، همیشه کپی جدید می‌سازد
# English: Flattens the array into 1D and always returns a copy
f = a.flatten()
 

# 3️⃣ ravel
# فارسی: تبدیل آرایه به 1بعدی، view ممکن است بازگردانده شود
# English: Flattens the array into 1D but returns a view if possible
r = a.ravel()
 

# 4️⃣ transpose
# فارسی: جابجایی سطر و ستون‌ها
# English: Swaps rows and columns of the array
t = a.T

# 5️⃣ resize
# فارسی: تغییر شکل آرایه؛ اگر سایز جدید بزرگ‌تر باشد، عناصر تکرار می‌شوند. 
# English: Changes the shape of an array; if the new size is larger, elements are repeated.
a = np.array([1, 2, 3, 4, 5, 6])

r1 = np.resize(a, (2, 3))   # همان تعداد عنصر
r2 = np.resize(a, (3, 4))   # بزرگ‌تر → تکرار عناصر
r3 = np.resize(a, (2, 2))   # کوچک‌تر → برش داده می‌شود

 



### reshape()
- Returns a view when possible  
- `arr.reshape(shape)`  
- Changes affect original data  

### resize()
- Always creates an independent copy  
- `np.resize(arr, shape)`  
- Changes don't affect original data

**np.logical_**
 
NumPy provides several element-wise logical functions to perform boolean operations on arrays.  
Common functions are:  
- `np.logical_and(x, y)` : Element-wise AND  
- `np.logical_or(x, y)`  : Element-wise OR  
- `np.logical_not(x)`    : Element-wise NOT  
- `np.logical_xor(x, y)` : Element-wise XOR

 

 

```python
 

# Boolean arrays
a = np.array([True, False, True, False])
b = np.array([True, True, False, False])

print("AND :", np.logical_and(a, b))  # [ True False False False]
print("OR  :", np.logical_or(a, b))   # [ True  True  True False]
print("NOT a:", np.logical_not(a))    # [False  True False  True]
print("XOR :", np.logical_xor(a, b))  # [False  True  True False]

# Numerical arrays with conditions
x = np.array([1, 2, 3, 4, 5, 6])

# Elements between 2 and 5
mask = np.logical_and(x > 2, x < 5)
print("Between 2 and 5:", x[mask])  # [3 4]

# Elements less than 2 or greater than 5
mask2 = np.logical_or(x < 2, x > 5)
print("Less than 2 or greater than 5:", x[mask2])  # [1 6]

# Using logical_not
print("Not greater than 3:", x[np.logical_not(x > 3)])  # [1 2 3]

1️⃣ np.take
**What it does:**  
Select specific elements by **index** along an axis. Useful for **reordering data or picking rows/columns**. <br>
انتخاب عناصر خاص با **ایندکس** در امتداد یک محور. برای مرتب‌سازی مجدد داده‌ها یا انتخاب ردیف‌ها / ستون‌ها مفید است

````python
data = np.array([[10, 20, 30],
                 [40, 50, 60],
                 [70, 80, 90]])
reordered = np.take(data, [2,0,7])
print(reordered)
````

````python
selected_rows = np.take(data, [0,2], axis=0)
print(selected_rows)
````

 🔹 `np.fromfunction(function, shape, dtype=float)`

 Syntax:
**np.fromfunction**(function, shape, dtype=float)

Parameters:
- **function**: A lambda or function that receives indices (i, j, k, ...)  
  Example: `lambda i, j: i + j`

- **shape**: Tuple defining the shape of the array  
  Example: `(3, 3)`, `(4, 5)`, `(2, 3, 4)`

- **dtype**: Data type of output (int, float, bool, ...)

---

📝 Common Patterns:

1. **Sum of indices** → `λ i, j: i + j`  
   👉 Creates a matrix where each element = row + col

2. **Multiplication of indices** → `λ i, j: i * j`  
   👉 Multiplication table

3. **Identity matrix** → `λ i, j: i == j`  
   👉 Equivalent to `np.eye(n)`

4. **Mathematical functions** → `λ i, j: np.sin(i) + np.cos(j)`

5. **Higher dimensions** → `λ i, j, k: i + j + k`  
   👉 Works for N-D arrays

Examples:

```python
# Sum of indices (3x3)
arr = np.fromfunction(lambda i, j: i + j, (3, 3))
# [[0 1 2]
#  [1 2 3]
#  [2 3 4]]

# Multiplication table
mult_table = np.fromfunction(lambda i, j: i * j, (5, 5), dtype=int)

# Identity matrix
identity = np.fromfunction(lambda i, j: i == j, (4, 4), dtype=int)

# Mathematical pattern
pattern = np.fromfunction(lambda i, j: np.sin(i) + np.cos(j), (3, 4))
```

---

### ⚡ Tips:
- Indices are floats by default → set `dtype=int` if needed
- Very useful for generating structured test arrays
- Great for quick matrix generation (like grids, eye, etc.)

📌 Syntax: <br>
**np.argsort**(arr, axis=-1)   → returns indices that would sort the array <br>

📌 Parameters: <br>
axis=None → flatten array <br> 
axis=0    → sort each column <br>
axis=1    → sort each row

````python
arr = np.array([40, 10, 30, 20])
indices = np.argsort(arr)      # اندیس‌هایی که آرایه را صعودی مرتب می‌کنند
print(arr[indices])            # [10 20 30 40]

# مرتب‌سازی نزولی
indices_desc = np.argsort(arr)[::-1]
print(arr[indices_desc])       # [40 30 20 10]

# مرتب کردن آرایه مرتبط با نمرات
scores = np.array([70, 90, 50, 80])
names = np.array(['Ali', 'Sara', 'John', 'Mary'])
sorted_names = names[np.argsort(scores)]
print(sorted_names)            # ['John' 'Ali' 'Mary' 'Sara']
````

 NumPy Sort Functions Reference

📌 `np.sort(a, axis=-1, kind='quicksort', order=None)`

- **Returns**: sorted array
- **axis=-1** → last axis (default)
- **kind** → `'quicksort'`, `'mergesort'`, `'heapsort'`, `'stable'`
- **order** → for structured arrays

````python
# Basic sorting
arr = np.array([3,1,4,2])
np.sort(arr)  # [1 2 3 4]

# Matrix sorting
matrix = np.array([[3,1,2],[6,4,5]])
np.sort(matrix, axis=1)  # row-wise: [[1 2 3],[4 5 6]]

# Structured array sorting
data = np.array([(3,'b'),(1,'a'),(2,'c')],
                dtype=[('x',int),('y','U1')])
np.sort(data, order='x')  # [(1,'a'),(2,'c'),(3,'b')]
````

- **choices**: list of values/arrays corresponding to conditions
- **default**: value if no condition is True

```python 
import numpy as np

a = np.array([1, 2, 3, 4, 5])
b = np.array([5, 4, 3, 2, 1])

# 1️⃣ np.add
# English: Adds two arrays element-wise
add_result = np.add(a, b)  # [6 6 6 6 6]

# 2️⃣ np.multiply
# English: Multiplies two arrays element-wise
mul_result = np.multiply(a, b)  # [5 8 9 8 5]

# 3️⃣ np.divide
# English: Divides two arrays element-wise
div_result = np.divide(a, b)  # [0.2 0.5 1. 2. 5.]

# 4️⃣ np.power
# English: Raises elements of first array to powers from second array
pow_result = np.power(a, b)  # [1 16 27 16 5]

# 5️⃣ np.mod
# English: Element-wise modulo
mod_result = np.mod(a, b)  # [1 2 0 0 0]

# 6️⃣ np.cumsum
# English: Cumulative sum
cumsum_result = np.cumsum(a)  # [1 3 6 10 15]

# 7️⃣ np.cumprod
# English: Cumulative product
cumprod_result = np.cumprod(a)  # [1 2 6 24 120]

# 8️⃣ np.diff
# English: Calculates the difference between consecutive elements
diff_result = np.diff(a)  # [1 1 1 1]
 

# 🔟 np.round / np.floor / np.ceil
# English: Rounds numbers to nearest integer, floor or ceil
arr = np.array([1.2, 2.5, 3.7])

round_result = np.round(arr)  # [1. 2. 4.]  
floor_result = np.floor(arr)  # [1. 2. 3.]   
ceil_result  = np.ceil(arr)   # [2. 3. 4.]
