<p dir="rtl">
📊 ۲. Axes چیه؟<br>
Axes در واقع خودِ نمودار است.<br>
اون جعبه‌ای که محورهای X و Y داخلش هست.<br>
Axes یعنی کل فضای نمودار (هم داده، هم برچسب‌ها، هم عنوان)<br>
Axis یعنی یکی از محورهای X یا Y داخل Axes
</p>

<p dir="rtl">
مثلاً وقتی می‌نویسی:<br>
<code dir="ltr">fig, ax = plt.subplots()</code><br>
در واقع داری هم Figure می‌سازی، هم یه Axes داخلش.
</p>

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [10, 20, 25])
plt.show()
```

<p dir="rtl">
📌 الان <code dir="ltr">ax</code> همون "Axes" است — یعنی یه نمودار روی Figure
</p>

&nbsp;

<p dir="rtl">
  رابطه‌ی Figure و Axes<br><br>
  می‌تونی این ساختار رو مثل یه درخت در نظر بگیری:<br>
  <code dir="ltr">Figure</code><br>
  &nbsp;├── <code dir="ltr">Axes 1</code><br>
  &nbsp;│&nbsp;&nbsp;&nbsp;&nbsp;├── <code dir="ltr">Axis X</code><br>
  &nbsp;│&nbsp;&nbsp;&nbsp;&nbsp;└── <code dir="ltr">Axis Y</code><br>
  &nbsp;└── <code dir="ltr">Axes 2</code> (اختیاری)<br><br>
  📍 یعنی:<br><br>
  یک Figure می‌تونه چند Axes داشته باشه (مثلاً برای subplotها)<br>
  هر Axes محورهای خودش رو داره
</p>

```python
fig , axs = plt.subplots(2,2, figsize = (10,8))

axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 1].bar([1, 2, 3], [3, 2, 1])
axs[1, 0].scatter([1, 2, 3], [1, 4, 9])
axs[1, 1].pie([10, 20, 30])

plt.show()
```

<p dir="rtl">
  در Matplotlib دو سبک برنامه‌نویسی اصلی وجود داره:<br>
  🔹 API سطح بالا (State-based / Pyplot interface)<br>
  🔹 API شی‌ءگرا (Object-oriented interface)<br><br>
  هر دو کار مشابه انجام می‌دن، ولی از نظر ساختار و کنترل تفاوت‌های مهمی دارن.
</p>

```python
import numpy as np 
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(6,3))
plt.plot(x, y, label='sin(x)')
plt.title("Pyplot API")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```

<p  style="text-align: center;">مثال</p>

```python
import seaborn as sns
# آماده کردن داده نمونه
data = sns.load_dataset("iris")

# نمایش چند تم Seaborn و Matplotlib
styles = [
    ("default", None),
    ("seaborn-darkgrid", lambda: sns.set_style("darkgrid")),
    ("seaborn-whitegrid", lambda: sns.set_style("whitegrid")),
    ("ggplot (matplotlib)", lambda: plt.style.use("ggplot"))
]

for name, setter in styles:
    # تنظیم استایل
    if setter:
        setter()
    else:
        plt.style.use("default")
    # رسم
    plt.figure(figsize=(5,2.5))
    sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species')
    plt.title(f"Style: {name}")
    plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, label="A")
ax.plot(x, y2, label="B")
ax.plot(x, y3, label="C")

handles, labels = ax.get_legend_handles_labels()

# فقط A و C نمایش داده شوند
select = [0, 2]
ax.legend([handles[i] for i in select],
          [labels[i] for i in select],
          title="Elective")

plt.show()
```

<p dir="rtl">

```python
fig, ax = plt.subplots()
ax.plot(x, y1, label="A")
ax.plot(x, y2, label="B")
ax.plot(x, y3, label="C")

handles, labels = ax.get_legend_handles_labels()

# ساخت دیکشنری از برچسب‌ها
legend_dict = dict(zip(labels, handles))

# ترتیب دلخواه بر اساس نام‌ها
desired_order = ["C", "A", "B"]
ax.legend([legend_dict[name] for name in desired_order],
          desired_order)

plt.show()
```

<p dir="rtl">
🎯 مفهوم کلی "Spine"<br><br>
در Matplotlib و Seaborn، هر محور (Axes) ۴ تا spine (لبه یا قاب) داره:<br><br>
<code dir="ltr">'top'</code> → قاب بالایی<br>
<code dir="ltr">'right'</code> → قاب سمت راست<br>
<code dir="ltr">'bottom'</code> → قاب پایینی (محور X)<br>
<code dir="ltr">'left'</code> → قاب سمت چپ (محور Y)<br><br>
📘 این خطوط معمولاً رنگ مشکی کم‌رنگ هستن و محورهای X و Y رو احاطه می‌کنن.
</p>

    <tr>

      <td>bool</td>
      <td>حذف قاب بالایی (پیش‌فرض <code dir="ltr">True</code>)</td>
    </tr>
    <tr>
      <td><code dir="ltr">right</code></td>
      <td>bool</td>
      <td>حذف قاب راست (پیش‌فرض <code dir="ltr">True</code>)</td>
    </tr>

      <td>bool</td>
      <td>حذف قاب چپ</td>
    </tr>

    </tr>
    <tr>
      <td><code dir="ltr">trim</code></td>
      <td>bool</td>

    </tr>

<p dir="rtl">
✅ مثال ۱: حذف همه قاب‌ها<br>
<code dir="ltr">sns.despine(top=True, right=True, left=True, bottom=True)</code><br><br>
✅ مثال ۲: فقط حذف قاب راست<br>
<code dir="ltr">sns.despine(right=True, top=False)</code><br><br>
✅ مثال ۳: کوتاه کردن قاب‌ها تا محدوده داده<br>
<code dir="ltr">sns.despine(trim=True)</code><br><br>
📘 در حالت <code dir="ltr">trim=True</code>، قاب‌ها فقط تا محدوده داده رسم می‌شن<br>
(نه تا انتهای شکل).<br><br>
✅ مثال ۴: فاصله دادن قاب از محور داده‌ها<br>
<code dir="ltr">sns.despine(offset=10)  # فاصله‌ی قاب از داده‌ها</code><br><br>
📘 معمولاً در نمودارهای ستونی یا violinplot مفیده.
</p>

```python
tips = sns.load_dataset("tips")

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# حالت معمولی
sns.set_style("whitegrid")
sns.scatterplot(data=tips, x="total_bill", y="tip", ax=axes[0])
axes[0].set_title("without despine")

# حالت با despine
sns.scatterplot(data=tips, x="total_bill", y="tip", ax=axes[1])
sns.despine(ax=axes[1])
axes[1].set_title("with sns.despine()")

plt.tight_layout()
plt.show()
```

<p dir="rtl">
🎯 ۱. مفهوم کلی<br>
<code dir="ltr">plt.tight_layout(pad=1.08, h_pad=None, w_pad=None, rect=None)</code><br><br>
تابع <code dir="ltr">plt.tight_layout()</code> به Matplotlib می‌گه:<br><br>
“به‌صورت خودکار فاصله‌ی بین عناصر شکل (نمودارها، عنوان‌ها، برچسب‌ها و محور‌ها) رو تنظیم کن تا چیزی روی هم نیفته.”<br><br>
🧩 ۲. چرا لازم داریمش؟<br><br>
به‌طور پیش‌فرض، Matplotlib همیشه فاصله‌ها رو خوب تنظیم نمی‌کنه.<br>
مثلاً در نمودارهای چندتایی (با <code dir="ltr">plt.subplot</code> یا <code dir="ltr">plt.subplots</code>) معمولاً برچسب محور یا عنوان نمودار با نمودار کناری یا پایینش برخورد می‌کنه 😬
</p>

<p dir = "rtl">به طور مثال</p>

```python
plt.subplot(2, 1, 1)
plt.title( "above chart")
plt.plot([1, 2, 3])

plt.subplot(2, 1, 2)
plt.title( " below chart")
plt.xlabel("X")
plt.plot([3, 2, 1])

plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.title( "above chart")
plt.plot([1, 2, 3])

plt.subplot(2, 1, 2)
plt.title( " below chart")
plt.xlabel("X")
plt.plot([3, 2, 1])

plt.tight_layout()
plt.show()
```

<p dir="rtl">
تفاوت <code dir="ltr">plt.tight_layout()</code> با <code dir="ltr">fig.tight_layout()</code><br><br>
وقتی از interface ساده (plt) استفاده می‌کنی، <code dir="ltr">plt.tight_layout()</code> کافیه.<br><br>
وقتی با Figure و Axes به صورت شیء‌گرا کار می‌کنی:<br><br>
<code dir="ltr">fig, ax = plt.subplots(2, 2)</code><br>
<code dir="ltr">fig.tight_layout(pad=2.0)</code>
</p>

<p dir="rtl">
نکته حرفه‌ای: جایگزین مدرن‌تر → constrained_layout<br><br>
از نسخه‌های جدید Matplotlib، روش پیشنهادی‌تر برای تنظیم فضاها اینه که موقع ساخت figure بنویسی:<br><br>
<code dir="ltr">fig, ax = plt.subplots(2, 2, constrained_layout=True)</code><br><br>
📘 این کار تقریباً همون کار tight_layout() رو انجام می‌ده ولی دقیق‌تر و هوشمندتر است (مخصوصاً با colorbarها).
</p>

```python
import matplotlib.pyplot as plt
import numpy as np

# داده‌ها
x = np.linspace(0, 10, 200)
y = np.sin(x) * 10 + 15  # دامنه بین 5 تا 25

# ایجاد شکل و محور
fig, ax = plt.subplots(figsize=(10,6))

# رسم نمودار اصلی
ax.plot(x, y, color='black', label='Signal')

# --- نواحی با axhspan و axvspan ---
ax.axhspan(10, 20, color='green', alpha=0.2, label='Normal range (y)')  # ناحیه افقی
ax.axvspan(2, 4, color='orange', alpha=0.2, label='Highlighted range (x)')  # ناحیه عمودی

# --- خطوط با axhline و axvline ---
ax.axhline(y=12, color='blue', linestyle='--', linewidth=2, label='Lower threshold (y)')
ax.axhline(y=18, color='red', linestyle='--', linewidth=2, label='Upper threshold (y)')
ax.axvline(x=3, color='purple', linestyle='-.', linewidth=2, label='Important point (x)')

# تنظیمات نمودار
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.set_title(' axhspan, axvspan, axhline, axvline')
ax.legend(loc='upper right')
ax.grid(True, linestyle=':', alpha=0.5)

plt.show()
```

<code dir="ltr"> plt.fill_between()</code>

<p dir="rtl">
برخلاف axhspan یا axvspan که بازه‌ها را روی محورهای X یا Y رنگی می‌کنند، fill_between می‌تواند مستقیماً بین دو سری داده یا منحنی‌ها ناحیه رنگی ایجاد کند و برای نمایش ناحیه خطا، محدوده اطمینان یا سایه‌دار کردن زیر نمودار بسیار کاربردی است.<br><br>
🎯 ۱. ساختار کلی<br>
<code dir="ltr">plt.fill_between(x, y1, y2=0, where=None, color=None, alpha=0.3, **kwargs)</code><br><br>
<code dir="ltr">x</code> → مختصات محور X<br>
<code dir="ltr">y1</code> → مختصات بالا یا منحنی اصلی<br>
<code dir="ltr">y2</code> → مختصات پایین یا منحنی مرجع (اگر حذف شود، پیش‌فرض 0 است)<br>
<code dir="ltr">where</code> → شرط برای رنگی کردن (مثلاً فقط بخش‌هایی که True هستند)<br>
<code dir="ltr">color</code> → رنگ ناحیه<br>
<code dir="ltr">alpha</code> → شفافیت (۰ تا ۱)<br>
<code dir="ltr">**kwargs</code> → تنظیمات اضافی مانند label برای legend
</p>

```python
x = np.linspace(0, 10, 200)
y = np.sin(x) * 10 + 15

plt.plot(x, y, color='black', label='Signal')

# رنگی کردن ناحیه بین نمودار و محور X
plt.fill_between(x, y, color='green', alpha=0.2, label='Area under curve')

plt.legend()
plt.show()
```

```python
y1 = np.sin(x) * 10 + 15
y2 = np.sin(x) * 5 + 10

plt.plot(x, y1, color='blue', label='Curve 1')
plt.plot(x, y2, color='red', label='Curve 2')

# ناحیه بین دو منحنی
plt.fill_between(x, y1, y2, color='purple', alpha=0.3, label='Area between curves')

plt.legend()
plt.show()
```

```python
plt.plot(x, y1, color='black')

# فقط بخش‌هایی که y1 > 15 هستند
plt.fill_between(x, y1, 15, where=(y1>15), color='orange', alpha=0.3)

plt.show()
```

```python
names = ['Ali', 'Sara', 'Reza', 'Neda']
scores = [75, 85, 60, 90]

# مرتب‌سازی بر اساس مقدار
sorted_idx = np.argsort(scores)
names = np.array(names)[sorted_idx]
scores = np.array(scores)[sorted_idx]

plt.barh(names, scores, color='orange')
plt.gca().invert_yaxis()   # تا بیشترین مقدار بالا قرار گیرد
plt.title("Sorted barh() with invert_yaxis()")
plt.xlabel("Score")

# برچسب عددی روی میله‌ها
bars = plt.barh(names, scores, color='orange')
plt.bar_label(bars, fmt='%.0f', padding=3)

plt.show()
```
