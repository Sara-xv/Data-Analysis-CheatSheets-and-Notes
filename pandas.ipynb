Generate descriptive statistics that summarize the central tendency, dispersion, and shape of a dataset's distribution.

* Analyzes **numeric columns** by default (ignoring text).
* Provides key metrics:
  * `count` → number of non-null observations
  * `mean` → average of the values
  * `std` → standard deviation (measure of dispersion)
  * `min` / `max` → minimum and maximum values
  * `25%`, `50%`, `75%` → percentiles (quartiles)
* Use `include` or `exclude` to specify data types (e.g., `'all'`, `'object'`, `np.number`).
* Use `percentiles` to customize the list of percentiles calculated.

Select data by labels (row and column names).
* Can select **single row/column** or **multiple rows/columns**.
* Returns a **Series** (for single column/row) or **DataFrame** (for multiple rows/columns).
* Slices are inclusive of the end label.

```python
df.loc['r3'] # access row by label
df.loc[:,'age'] # access column by label
df.loc['r1', 'age'] # access cell by label
df.loc[['r1','r2']] # access multiple rows by label
df.loc[:,['age','city']] # access multiple columns by label
df.loc[['r1','r3'], ['name','city']] # access multiple rows and columns by label
df.loc['r2':'r4' , ['name' , 'city']] # access range of rows and specific columns by label
df.loc[data['age'] > 25] # filter rows based on condition
df.loc['r1' , 'age'] = 47 # modify cell value by label
df.loc['r3' , ['name' , 'age']]  = ["Maryam" , 28] # modify multiple cell values by label
```

Select data by **integer positions** (row and column indices).

* Can select **single row/column** or **multiple rows/columns**.
* Returns a **Series** (for single column/row) or **DataFrame** (for multiple rows/columns).
* Slices are **exclusive of the end index**.


```python
df.data.iloc[0] # access first row  
df.iloc[:,0] # access first column
df.iloc[1,2] # access cell at row 1, column 2
df.iloc[[0,2]] # access multiple rows by integer location
df.iloc[:, [0,2]] # access multiple columns by integer location
df.iloc[1:4, [0,2]] # access range of rows and specific columns by integer location
df.iloc[-1] # access last row
df.iloc[:, -1] # access all rows except last row
df.iloc[df['age'] > 25] # filter rows based on condition using integer location
df.iloc[1, [1, 2]] = [32, "Karaj"] # modify multiple cell values  
df.iloc[[0, 2], [1, 2]] = [[28, "Qom"], [23, "Urmia"]] # modify multiple rows and columns
```

Fast access to a single value in a DataFrame

* **`.at`**: Access a value using **row and column labels**.
* **`.iat`**: Access a value using **row and column integer positions**.


```python
df.at["r2", "age"] # access cell by label (faster for single value)
df.at["r3", ["name", "city"]] = ["Lina", "Shiraz"] # modify multiple cell values by label
df.iat[1,2] # access cell by integer location (faster for single value)
df.iat[2,1] = 29 # modify cell value by integer location
```

 Fill missing values (`NaN`) in a DataFrame or Series.

 

### **Parameters explained:**

* **`value`**:

  * The value used to replace missing values.
  * Can be a single value (applied to all NaNs) or a dictionary specifying different values for each column.

* **`method`**:

  * Method for filling missing values.
  * Options:

    * `'ffill'` or `'pad'` → forward fill (use previous valid value)
    * `'bfill'` or `'backfill'` → backward fill (use next valid value)

* **`axis`**:

  * Determines the axis along which to fill.
  * `0` or `'index'` → fill column-wise
  * `1` or `'columns'` → fill row-wise

* **`inplace`**:

  * If `True`, fill values **in the original DataFrame**.
  * If `False` (default), returns a **new DataFrame**.

* **`limit`**:

  * Maximum number of consecutive NaNs to fill.
  * Useful with `method='ffill'` or `bfill`.




```python
df.fillna(0) # fill NaN values with 0
df.fillna(0,axis=1) # fill NaN values with 0 along columns
df.fillna(0,axis=0)  # fill NaN values with 0 along rows
df.fillna(method='ffill') # forward fill NaN values
df.fillna(method='bfill') # backward fill NaN values
df.fillna(df.mean()) # fill NaN values with column mean
df['age'].fillna(df['age'].mean()) # fill NaN values in 'age' column with its mean
df.fillna({'age': df['age'].mean(), 'city': 'Unknown'}) # fill NaN values with different strategies per column
```

```python
# WARNING: Using Series.fillna(method='ffill' or 'bfill') is deprecated in pandas and will raise an error in future versions.
df['age'] = df['age'].ffill()
df['age'] = df['age'].bfill()
```

Column(s) to use as column labels (keys to group by).

 **• `aggfunc`**
Function(s) to use for aggregating (default: 'mean'). e.g., 'sum', 'count', 'len'.

 **• `fill_value`**
Value to replace missing values (NaN) with.

 **• `margins`**
* `False` → do not add all row/column (default)
* `True` → add row/column for grand totals
  
 **• `dropna`**
* `True` → do not include columns whose entries are all NaN
* 
 **• `margins_name`**
Name of the row/column that will contain the totals (default: 'All').
 

```python
pd.pivot_table(df, values='Sales', index='Date', columns='Product', aggfunc='sum') # basic sum
pd.pivot_table(df, index='Region', columns='Product', aggfunc='mean') # average (default)
pd.pivot_table(df, values='Price', index='Category', aggfunc=['min', 'max']) # multiple functions
pd.pivot_table(df, index='Month', columns='Year', fill_value=0) # replace NaN with 0
pd.pivot_table(df, index='Store', margins=True) # add grand totals
pd.pivot_table(df, aggfunc={'Sales': 'sum', 'Qty': 'mean'}) # different agg per column
# or 
df.pivot_table(index='animal', columns='visits', values='age', aggfunc='mean')
```

 **• `axis`**

* 
 **• `fill_value`**

 **• `level`**
